### Now let's load the facies

```python
header = ['SitID','Depth','LithID','W_Tar','SW','VSH','PHI','RW']
```

header details:

- SitID	A number used within AGS to identify wells.
- Depth	The depth of an interval in metres from the Kelly Bushing elevation.
- LithID	A code to identify the lithology.
- W_Tar	Mass percent bitumen.
- SW	Water saturation.
- VSH	Volume of shale.
- PHI	Density.
- RW


```python
path_facies_data = "../data/Mannville_input_data/v0.0.0-alpha/OilSandsDB/INTELLOG.TXT"


facies_data = pd.read_csv(path_facies_data, sep="\t",header = None)
facies_data.columns = header
```

```python
facies_data.head()
```

```python
len(facies_data)
```

```python
new_df = pd.merge(df_all_wells_wKNN, facies_data,  how='left', left_on=['SitID','DEPT'], right_on = ['SitID','Depth'])
```

```python
new_df3 = new_df2[['CALI', 'COND', 'DELT', 'DEPT', 'DPHI', 'DT', 'GR', 'ILD', 'ILM',
       'NPHI', 'PHID', 'RHOB', 'SFL', 'SFLU', 'SN', 'SP', 'UWI',
       'SitID', 'lat', 'lng',
       'Depth', 'LithID', 'W_Tar', 'SW', 'VSH', 'PHI', 'RW']]
```

```python
#new_df3["lithName"] =  lith_conversion[str(new_df3.LithID)]

new_df3["lithName"] = new_df3["LithID"].map(lith_conversion )
```
